Project Title:  Udemy Courses Data Analysis

Project Description
The Udemy Courses Data Analysis project focuses on exploring and analyzing Udemy course data to understand trends in online education. The main purpose of this project is to study course subjects, pricing, number of subscribers, reviews, and publishing timelines to gain meaningful insights about learner behavior and course performance.

Using Python and data analysis libraries such as pandas, numpy, matplotlib, and seaborn, the dataset is cleaned and processed to remove inconsistencies and prepare it for analysis. The published_timestamp column is converted into datetime format, and new features such as year are extracted to study publishing trends over time.


The analysis focuses on:

1. Course and Pricing Analysis
This analysis focuses on comparing paid and free courses available on Udemy. It examines the price distribution of paid courses and studies how course pricing impacts the number of subscribers and reviews. The goal is to understand whether pricing plays a significant role in course popularity and learner engagement.

2. Subject-wise Performance
This section analyzes the performance of different course subjects on Udemy. It includes the number of courses available in each subject, the average number of subscribers and reviews per subject, and identifies the most popular course categories based on learner interest.

3. Subscriber and Review Insights
This analysis explores the relationship between the number of subscribers and the number of reviews received by courses. It identifies top-performing courses and examines engagement patterns across different subjects to understand how learners interact with course content.

4. Publishing Trends
This section focuses on course publishing patterns over time. It analyzes the number of courses published each year, studies the growth of Udemy’s content, and uses publishing timestamps to identify trends in course creation.


Problem Statement:

With the rapid growth of online learning platforms like Udemy, it is important to understand course performance and learner preferences. However, analyzing large amounts of course data manually is difficult. This project aims to analyze Udemy course data to identify popular subjects, compare paid and free courses, study subscriber and review trends, and understand course publishing patterns.



Objective:

The objective of this project is to analyze Udemy course data and provide insights that help:

Identify high-demand subjects on Udemy.
Compare the performance of paid and free courses.
Understand learner engagement through subscribers and reviews.
Analyze course publishing trends over time.


Outcome:

The outcome of this project is a set of clear insights and visualizations that explain Udemy course trends. The analysis helps in understanding learner preferences, course popularity, and growth of online education over time.
